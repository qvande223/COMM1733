```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
# Load data
df = pd.read_csv('video_comment_likes (1).csv')
```

```python
# Sort the data by 'likes' column in descending order
df_sorted = df.sort_values('Likes', ascending=False)
```

```python
# Select the top N comments based on the desired number
top_n = 10  # Set the desired number of top comments
top_comments = df_sorted.head(top_n)
```

```python
# Create a bar plot of the top liked comments
plt.figure(figsize=(10, 6))
plt.barh(top_comments['Comment'], top_comments['Likes'], color='green')
plt.xlabel('Likes')
plt.ylabel('Comment')
plt.title('Top Liked Comments')
plt.gca().invert_yaxis()  # Invert the y-axis to display comments from top to bottom
plt.show()
```
